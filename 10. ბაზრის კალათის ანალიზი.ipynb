## ბაზრის კალათის ანალიზი

მონაცემთა მეცნიერებას და მანქანურ დასწავლას მრავალ სფეროში აქვს გამოყენება. ერთერთი მათგანია საცალო გაყიდვების ბიზნესი. წარმოვიდგინოთ რომ ვართ საცალო გაყიდვების ბიზნესის მფლობელები სადაც ასობით ნივთს ვყიდით. ყოველ თვე ასობით ტრანზაქცია ხორციელდება ჩვენს მაღაზიაში. თითოეული ტრანზაქციის დროს ერთზე მეტი ნივთის შეძენა ხდება. ეს ნიშნავს რომ ერთზე მეტი ნივთია თითოეულ ტრანზაქციაში.

თვის ბოლოს გვინდა რომ გავზარდოთ გაყიდვები ჩვენს მაღაზიაში, ამისთვის კი ვცდილობთ  გავარკვიოთ თუ როგორი პრომო აქციები და ფასდაკლებები შევთავაზოთ კონკრეტულ ნივთებზე მომხმარებლებს. შესაბამისად ვსვამთ კიტხვებს თუ რომელ პროდუქტზე უნდა გვქონდეს ფასდაკლება, არის თუ არა რაიმე კანონზომიერება ჩვენს გაყიდვებში, რათა შემდგომ პრომო აქციები უფრო ზუსტად და ეფექტურად გამოვიყენოთ.

ამ კითხვაზე პასუხის გასაცემად ჩვენ გამოვიყენებთ მონაცემთა მეცნიერებაში და მანქანურ დასწავლაში კარგად ცნობილ მეთოდს Market Basket Analysis–ის.
ეს მეთდი გვეხმარება დავინახოთ კანონზომიერებები თუ ზოგიერთი ნივთის გაყიდვა როგორ გავლენას ახდენს სხვა ნივთების გაყიდვაზე.
მსგავსი კავშრების პოვნა საცალო გამყიდველებს ეხმარებათ მარკეტინგული სტრატეგიების დაგეგმვაში.

MBA არის ასოციაციის ტიპის ამოცანა. ასოციაციის წესი იწერება შემდეგნაირად: A => B სადაც A არის antecedent ხოლო B არის consequent.
Support, confidence, და Lift არის სამი მნიშვნელოვანი ასოციაციის აღმოჩენის შეფასების კრიტერიუმი. 

Support

support–ის დონე არის თუ რამდენად ხშირად გვხვდება კომბინაცია market basket–ში (მონაცემებში). Support არის პროცენტი კალათების/ტრანზაქციების რომელიც შეიცავენ A და B–ს ასოციაციებს.


Confidence

ასოციაციის სიმძლავრე განისაზღვრება მისი Confidence–ით. რაც არის პროცენტი იმ შემთხვევების სადაც Confidence (B) გვხვდება იმპირობით რომ antecedent (A) -ს ქონდა ადგილი.  
Confidence არის იმის პროცენტი რომ კალათაში გვაქვს A რომელშიც გვაქვს B.

Confidence(A => B) ≠ Confidence(B => A).

შენიშვნა:

Expected confidence (A => B) = P(B)

Lift

Lift = Confidence/Expected confidence.

Lift არის B-ს კალათაში პოვნს ალბათობის იმ პირობით რომ კალათაში იქნება A,  შეფარდებული B-ს შემთხვევიტად კალათაში პოვნის ალბათობასთან



MBA-ს გადაჭრა ხდება "აპრიორი" ალგორითმის გამოყენებით. მოდით დაწვრილებით ავხსნათ ეს ალგორითმი.

შემოვიტანოთ კალათის მონაცემები:

1. ვნახოთ ტრანზაქციების რაოდენობა სადაც თითოეული ნივთი ხვდება:

2. გამოვიყენოთ მინიმალური suppot–ის წესი (დავტოვოთ 3 და მეტი ტრანზაქციები): 

3. დარჩენილი ნივთებით შევქმნად წყვილ წყვილად ყველა შესაძლო კომბინაცია: MO, MN, MK, ME, MY, ON, OK, OE, OY, NK, NE, NY, KE, KY, EY.

4. ვნახოთ თითოეული კომბინაცია რამდენჯერ არის ნაყიდი:

5. დავტოვოთ მინიმალურ support–ს ზემოთ არსებული წყვილები:

6. სამი ნივთის ერთად შესაკვრელად კიდევ ერთი წესი გვჭირდება. ზედა ცხრილის წყვილებიდან შევარჩით ისეთი წყვილები რომლების პირველი სიმბოლო ერთი და იგივეა, ასე რომ მივიღებთ OK და OE, რაც გვაძლევს OKE-ს. KE და KY გვაძლევს KEY–ს.


შემოვიტნაოთ ბირტანეთში ონლაინ საცალო გაყიდვების მონაცემები:

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('online_retail_II.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
#სიმარტივისთვის გადავყაროთ სიცარიელეები
df = df.dropna()
```

მონაცემებში, სვეტი 'Quantity' გვაჩვენებს თუ რამდენი ნივთი იქნა შეძენილი თითოული ტრანზაქციის დროს(InvoiceNo). რადგან ეს ონლაინ რითეილის მონაცემებია, ზოგჯერ ტრანზაქცია უქმნდეა და ამიტომ გვაქვს Quantity–ის მნიშვნელობები 0, რომლებიც უნდა გადავყაროთ.

```python
df_1 = df[df['Quantity']>=0]
df_1.shape
```

```python
df.Country.value_counts()
```

```python
#მოდით მონაცემები შევზღუდოთ ბრიტანეტზე, სიმარტივისტვის.
df_uk = df_1[df_1['Country']=='United Kingdom']
df_uk.shape
```

შევქმნათ "კალათის" მონაცემები. კალათის მონაცემები მოიცავს თითოეულ ტრანზაქციაზე გაყიდული ნივთის რაოდენობას:

```python
df_basket = (df_uk.groupby(['Invoice', 'Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('Invoice'))
```

```python
df_basket.head()
```

ფაქტიურად ეს მონაცემები არის მომხარებლის კალატა რითიც მიდიან სალაროსთან.

ეს გვაჩვენებს თუ რამდენი ნივთი კონკრეტული ნივთი იყიდა მომხმარებელმა. თუ მნიშნველობა არის 0 მაშინ მომხარებელს არ უყიდია რაიმე.

market basket analysis–ის დროს  თუ რამდენი ნივთი იყიდა მომხარებელმა არ არის მნიშვნელოვანი, ჩვენთვის მნიშვნელოვანია ნვთი იყიდა თუ არა მომხმარებელმა. ამიტომ 1 ზე მეტი მნიშვნელობები კალათის ცხრილში შევცვალოთ 1–ით:

```python
def encode_1(x):
    if x <=0:
        return 0
    if x >=1:
        return 1
    
df_basket1 = df_basket.applymap(encode_1)    
```

```python
df_basket1.head()
```

```python
df_basket1.shape
```

market basket analysis დროს ჩვენ უნდა ვნახოთ ასოციაციები 2 და ეტ გაყიდულ ნივთებს შორის, ისტორიული გაყიდვების მონაცემების მიხედვით. შესაბამისად ნაკლებად გამოსადეგია ისეთი ტრანზაქციები სადაც მხოლოდ ერთი ნივთია გაყიდული. შემდეგი ნაბიჯოა დავფილტროტ ჩვენი კალათის მონაცემები და შევხედოთ მხოლოდ ისეთ ტრანზაქციებს სადაც 1–ზე მეტი ნივთია გაყიდული:

```python
df_basket2 = df_basket1[(df_basket1 > 0 ).sum(axis = 1) >=2]
df_basket2.shape
```

მონაცემების დამუშავების შემდგომ შეგვიძლია გამოვიყენოთ "apriori" ალგორითმი. ეს ალგორითმი გამოიყენება ხშირად ნაყიდი ნივთების დასადგენად. 


"apriori" ალგორითმის გამოსაყენებლად  ჩავწეროთ ბიბლიოთეკა  “mlxtend”.

```python
#MBA–ს ბიბლიოთეკა
# !pip install mlxtend
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
frequent_itemsets_plus = apriori(df_basket2, min_support=0.03, 
                                 use_colnames=True).sort_values('support', ascending=False).reset_index(drop=True)

frequent_itemsets_plus['length'] = frequent_itemsets_plus['itemsets'].apply(lambda x: len(x))

frequent_itemsets_plus
```

"apriori" ალგორითმის გამოყენებით ჩვენ შეგვიძლია განვსაზღვროთ სიხშირე, კერძოდ ჩვენს შემთხვევაში ჩვენ განვსაზღვრეთ ხშირად ნაყიდი ნივთები როგორც, ნივთები რომლებიც ნაყიდია მთლიანი ტრანზაქციიდან 3%–ზე მეტჯერ, შემდეგ დავამატეთ სვეტი 'length' სადაც მითითებულია ნაყიდი ნივთების რაოდენობა
როგორც ვხედავთ გვაქვს 108 ტრანზაქცია რომელიც ჩავთვალედ ხშირად ნაყიდად. (WHITE HANGING HEART T-LIGHT HOLDER) არის ყველაზე ხშირად ნაყიდი ნივთი.

apriori ალგორითმის გამოყენებით ხშირად ნაყიდი ნივთების პოვნის შემდგომ, ჩვენ უნდა გამოვიყენოთ ასოციაციის წესები, საიდანაც ამოვიღებთ ინფორმაციას თუ რომელი ნივთები იყიდება ერთად. ერს არის ჩვენი ამოცანის მთლიანი აზრი.

```python
from mlxtend.frequent_patterns import association_rules
```

```python
association_rules(frequent_itemsets_plus, metric='lift', 
                  min_threshold=1).sort_values('lift', ascending=False).reset_index(drop=True)
```

association_rules–ის შედეგებიდან ჩვენ ვხედავთ რომ  ROSES REGENCY TEACUP AND SAUCER და GREEN REGENCY TEACUP AND SAUCER ნივთებს გააჩნიათ ყველაზე მარალი ასოციაცია ერთმანეტთან, ვინაიდან ამ ორს აქვთ ყველაზე მაღალი “lift”–ის მნიშვნელობა. რამდენადაც მეტია ეს მნშვნელობა იმდენად მეტია ასოციაცია ნივთებს შორის. თუ lift–ის მნიშვნელობა 1 –ზე მეტია შეგვიძლია ვტქვათ, რომ ეს ორი ნივთი ასოცირებულია ერტმანეთთან. ამ შემთხვევაში 17.71 ძალიან მაღალია. ეს ნიშნავს, რომ ამ ორი ნივთის გაყიდვა ერთად კარგი აზრია.

ამასთან ერტად  support value = 0.0309 ROSES REGENCY TEACUP AND SAUCER და GREEN REGENCY TEACUP AND SAUCER თვის, რაც ნიშნავს რომ 3.09% სრული ტრანზაქციების იყო ამ ორი ნივთის ერთად გაყიდვა. რიცხვებში 476 ჯერ.
