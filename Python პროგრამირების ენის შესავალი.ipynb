### განვიხილოთ კიდევ ერთი მარტივი ამოცანა: გვაქვს ლისტი [1,3,12,5,33,12,15] , ლისტი შევცვალოთ ისე, რომ თითოეულ კენტ რიცხვს დავუმატოთ ერთი, ხოლო ლუწები გავყოთ ორზე. შედეგად მივიღებთ [2,4,6,6,34,6,16]

```python
lst = [1,3,12,5,33,12,15]

for index, item in enumerate(lst):
    if item % 2 == 0:
        lst[index] = lst[index]//2
    else:
        lst[index] = lst[index]+1
print(lst)
```

### ამოცანა 2: შეიყვანეთ რიცხვი n, და დაბეჭდეთ 0-დან n-მდე რიცხვების ჯამი(შემდეგ ცადეთ მხოლოდ კენტების, მხოლოდ ლუწების და ორივე ერთდროულად)

```python
#ამოხსნა ფუნქციის გარეშე, მარტივი მაგალითი
n = int(input())
sum = 0
for i in range(n):
    sum = sum+i
print(sum)
```

```python
#ამოხსნა ფუნქციით, ფუნქციის შექმნა
def get_sum(n):
    sum = 0
    for i in range(n):
        sum = sum+i
    return sum
```

```python
#n რიცხვის შეყვანა და ფუნქციის გამოძახება
n = int(input())
returned_sum = get_sum(n)
print(returned_sum)
```

### ვცადოთ კიდევ ერთი ამოცანის ამოხსნა. დავბეჭდოთ ყველა მარტივი რიცხვი 0-დან ჩვენს შეყვანილ რიცხვამდე და ეს მარტივი რიცხვები გამრავლებული ორზე. მაგალითად თუ შევიყვანთ 6-ს, უნდა დაგვიბეჭდოს 2,3,5 და 4,6,10
### იმისთვის რომ დავბეჭდო ყველა მარტივი რიცხვი, ჯერ გავაკეთოთ ფუნქცია რომელსაც გადავცემთ რიცხვს და გვეტყვის მარტივია თუ არა. ხოლო შემდეგ ჩავსვამთ ციკლში და შევამოწმებინებთ ყველა რიცხვს 0-დან ჩვენს შეყვანილ მნიშვნელობამდე

```python
def isPrime(n):
    for i in range(2,n):
        if n%i == 0:
            return False
    return True
```

```python
n = int(input())
for i in range(1,n):
    if isPrime(i)==True:
        print(i,"->",i*2)
```
